```python
import polars as pl
```

```python
df = pl.read_csv("../Hoiho_Genomes_24Feb2024_JGG_3Pops.csv", separator="\t")
df.head()
```

```python
# Only "Northern"
df_northern = df.filter(pl.col("Population3") == "Northern")
```

```python
df_gwas_samples = df_northern.select([pl.col("ID"), pl.col("RDS Status")])
df_gwas_samples = df_gwas_samples.filter(pl.col("RDS Status").is_in(["RDS", "No"]))
```

```python
# Replace RDS with 1 and No with 0
df_gwas_samples = df_gwas_samples.with_columns(
    pl.when(pl.col("RDS Status") == "RDS").then(1).otherwise(0).alias("RDS Status")
)
df_gwas_samples
```

```python
# Save
df_gwas_samples.write_csv("gwas_samples.csv", separator="\t")
```
