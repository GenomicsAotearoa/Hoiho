```python
# Individuals: 251 Populations: 3 Loci: 9999
# Population Index -> Population Label:
# 2->Campbell 1->Enderby 0->Northern
# Migration Rates:
# m[0][0]: 0.9932(0.0039) m[0][1]: 0.0023(0.0023) m[0][2]: 0.0045(0.0031)
# m[1][0]: 0.0184(0.0122) m[1][1]: 0.9631(0.0170) m[1][2]: 0.0184(0.0125)
# m[2][0]: 0.0485(0.0134) m[2][1]: 0.0087(0.0061) m[2][2]: 0.9428(0.0144)

# Make that a dataframe or something so we can make it into a heatmap
import numpy as np
import pandas as pd

migration_rates = np.array([
    [0.9932, 0.0023, 0.0045],
    [0.0184, 0.9631, 0.0184],
    [0.0485, 0.0087, 0.9428]
])

migration_df = pd.DataFrame(migration_rates, columns=["Northern", "Enderby", "Campbell"], index=["Northern", "Enderby", "Campbell"])
import plotly.express as px
fig = px.imshow(migration_df, text_auto=True, aspect="equal", color_continuous_scale="Viridis")
fig.update_layout(
    title="Migration Rates",
    xaxis_title="From Population",
    yaxis_title="To Population",
    width=800,
    height=600,
    coloraxis_colorbar=dict(title="Migration Rate")
)
fig.show()

pio.write_image(fig, "migration.svg", scale=6, width=1080, height=768)
```
