```python
import polars as pl
import plotly as plt
import plotly.express as px
import plotly.io as pio
```

```python
#eigenvec = pl.read_csv("removed_qual99_maf0.05.eigenvec", separator="\t")
# merged.a9.filtered.qual99_fmissing0.2.maf0.05.biallelic.eigenvec
eigenvec = pl.read_csv("merged.a9.filtered.qual99_fmissing0.2.maf0.05.biallelic.eigenvec", separator="\t")
eigenvec = eigenvec.rename(mapping={"#IID": "ID"})
metadata = pl.read_csv(
    "Hoiho_Genomes_Cleaned_Oct24_24.csv", separator="\t"
)

metadata = metadata.with_columns(
    pl.col("ID").replace("P29 ", "P29").alias("ID")
)

# Next is C101/CE9
metadata = metadata.with_columns(
    pl.col("ID").replace("C101/CE9", "CE9").alias("ID")
)

```

```python
metadata = metadata.with_columns(
    pl.when(pl.col("Location").is_in(["Unknown"]).not_())
    .then(pl.col("Location"))
    .otherwise(pl.lit("Northern Unknown"))
    .alias("Location")
)
```

```python
joined = eigenvec.join(metadata, on="ID", how="inner")
```

```python
# Load up the eigenval
eigenval = pl.read_csv(
    "removed_qual99_maf0.05.eigenval", separator="\t", has_header=False
)
# Convert eigenval to a simple list
eigenval = eigenval.to_pandas().iloc[:, 0].tolist()
```

```python
fig = px.scatter(joined, "PC1", "PC2", color="Location", template="plotly_white",
                 labels= {
                     "PC1": "PC1 (20.2% Variance)",
                     "PC2": "PC2 (10.8% Variance)",
                 },
                 width=1080,
                 height=768,
                )
fig.update_traces(marker_size=4)

#pio.write_image(fig, "all_locations.jpeg", scale=6, width=1080, height=768)
pio.write_image(fig, "all_locations.svg", scale=6, width=1080, height=768)
fig
```

```python
metadata = metadata.with_columns(
    pl.when(pl.col("Location").is_in(["Enderby", "Campbell"]))
    .then(pl.col("Location"))
    .otherwise(pl.lit("Mainland"))
    .alias("Location")
)
metadata
joined = eigenvec.join(metadata, on="ID", how="inner")
```

```python
fig = px.scatter(joined, "PC1", "PC2", color="Location", template="plotly_white",
                labels= {
                     "PC1": "PC1 (20.2% Variance)",
                     "PC2": "PC2 (10.8% Variance)",
                 },
                 width=1080,
                 height=768,
                )
fig.update_traces(marker_size=4)
#pio.write_image(fig, "general_locations.jpeg", scale=6, width=1080, height=768)
pio.write_image(fig, "general_locations.svg", scale=6, width=1080, height=768)

fig
```
